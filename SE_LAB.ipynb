```python
# Printing all even numbers, odd numbers, count of even numbers, count of odd numbers within a given range

def print_numbers_and_counts(start, end):
    even_count = 0
    odd_count = 0
    print("Even numbers:")
    for num in range(start, end + 1):
        if num % 2 == 0:
            print(num, end=" ")
            even_count += 1
    print("\nOdd numbers:")
    for num in range(start, end + 1):
        if num % 2 != 0:
            print(num, end=" ")
            odd_count += 1
    print("\nCount of even numbers:", even_count)
    print("Count of odd numbers:", odd_count)
start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))
print_numbers_and_counts(start_range, end_range)
```

```python
# Accepting 5 different subject marks from user and displaying the grade of the student.

a=int(input("Enter marks of the first subject: "))
b=int(input("Enter marks of the second subject: "))
c=int(input("Enter marks of the third subject: "))
d=int(input("Enter marks of the fourth subject: "))
e=int(input("Enter marks of the fifth subject: "))
avg=a+b+c+d+e/5
if(avg>=90):
    print("Grade: A")
elif(avg>=80&avg<90):
    print("Grade: B")
elif(avg>=65&avg<80):
    print("Grade: C")
elif(avg>=55&avg<70):
    print("Grade: D")
elif(avg>=45&avg<50):
    print("Grade: E")
else:
    print("Grade: F")
```

```python
# c)Generate Fibonacci series up to N numbers

def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
N =12
fibonacci(N)
```

```python
# Compute compound interest using loop for a certain principal and interest amount

principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the annual interest rate (in %): "))
time = int(input("Enter the number of years: "))
rate /= 100
amount = principal
for i in range(time):
  amount = amount * (1 + rate)
compound_interest = amount - principal
print("Final amount:", amount)
print("Compound interest:", compound_interest)
```

```python
def is_prime(digit):
  if digit < 2:
    return False
  for i in range(2, int(digit**0.5) + 1):
    if digit % i == 0:
      return False
  return True

number = input("Enter a number to find prime digits: ")
print("Prime digits in the number:")
for digit in number:
  if digit.isdigit():
    digit_int = int(digit)
    if is_prime(digit_int):
      print(digit)
```
